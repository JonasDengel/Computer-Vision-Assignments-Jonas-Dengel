```python
#Assignment 1:
import cv2
import time
import numpy

cap = cv2.VideoCapture(1)
start_time = time.time()
frame_count=0

while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blue,green,red = cv2.split(frame)
    #print(gray)
    frame_count += 1
    elapsed_time = time.time() - start_time
    fps = frame_count / elapsed_time
    print(fps)
    cv2.putText(frame, f"FPS: {fps:.2f}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    _, _, _, max_locg = cv2.minMaxLoc(gray)
    cv2.circle(frame,max_locg,10,(0,0,0),-1)
    
    
    red_mask = cv2.inRange(frame, (0,0,100), (50,50,255))
    max_logr = numpy.unravel_index(numpy.argmax(numpy.ma.masked_where(red_mask == 0, red)), red.shape)

  
    cv2.circle(frame, max_logr, 10,(0,0, 255), -1)


    #Part 6
    max_intensity = 0
    max_loc = (0, 0)

    for y in range(frame.shape[0]):
        for x in range(frame.shape[1]):
            intensity = gray[y, x]
            if intensity > max_intensity:
                max_intensity = intensity
                max_loc = (x, y)

    cv2.circle(frame, max_loc, 30, (0, 255, 0))
    
    
    
    cv2.imshow('frame', frame)
    #cv2.imshow('gray',gray)
    #cv2.imshow('red',red)
    
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()

cv2.destroyAllWindows()
```

```python
#Assignment 2
import cv2
import numpy as np
import random

cap = cv2.VideoCapture(0) 


def distance_to_line(x, y, x0, y0, vx, vy):
    numerator = np.abs((y - y0) * vx - (x - x0) * vy)
    denominator = np.sqrt(vx**2 + vy**2)
    distance = numerator / denominator
    return distance

while True:
    ret, frame = cap.read()
    edges = cv2.Canny(frame, 50, 150)
    i=0
    N=0
    Nmax=0
    limit=5
    
    epoints = np.column_stack(np.where(edges > 0))
    ep_s = epoints[::1]          
    while i<500:
        a, b = random.sample(range(len(ep_s)-1), 2)
        array = np.array([ep_s[a], ep_s[b]])
        
        vx, vy, x0, y0 = cv2.fitLine(array, cv2.DIST_L2, 0, 0.01, 0.01)
        
        N = np.sum(distance_to_line(ep_s[:, 0], ep_s[:, 1], x0, y0, vx, vy) < limit)
        
        if N > Nmax:
            best_line = (vx, vy, x0, y0)
            Nmax = N
            N=0
        i=i+1 
        
    x1, y1 = int(best_line[2] - 1000 * best_line[0]), int(best_line[3] - 1000 * best_line[1])
    x2, y2 = int(best_line[2] + 1000 * best_line[0]), int(best_line[3] + 1000 * best_line[1])

        
    cv2.line(frame, (y1, x1), (y2, x2), (0, 255, 0), 2)

    cv2.imshow('frame', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
